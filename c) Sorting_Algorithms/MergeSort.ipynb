# Mergesort

**We want to do the sorting within the array and not create another array : DO IT IN-PLACE Sorting**

```python
# In-place changes done in the array
def merge(a1, a2, a):
    
    i=0 ; j=0 ; k=0
    
    while i < len(a1) and j < len(a2):
        
        if a1[i] < a2[j]:
            a[k] = a1[i]
            k = k+1
            i = i+1
        else:
            a[k] = a2[j]
            k = k+1
            j = j+1
            
    while i < len(a1):
        a[k] = a1[i]
        k = k+1
        i = i+1
    
    while j < len(a2):
        a[k] = a2[j]
        k = k+1
        j = j+1
        
        
def mergeSort(arr):
    # Please add your code here
    
    if len(arr) == 0 or len(arr) == 1:
        return
    
    
    mid = len(arr)//2
    
    a1 = arr[0:mid]
    a2 = arr[mid:]
    
    # Induction Hypothesis : we assume that we will get our 2 halves sorted by calling mergesort on them
    
    # we are doing changes IN-PLACE by sending reference of the list
    # as these are in-place changes we do not need to return the array
    mergeSort(a1)
    mergeSort(a2)
    
    # Induction Step
    
    # the merging can be done IN-PLACE since arrays/lists are mutable 
    # therefore we are sending the references of the arrays/lists
    merge(a1, a2, arr)
    
# Main
n=int(input("Enter the size of the array to be sorted"))
arr=list(int(i) for i in input("Enter the elements of the array of size {} ::".format(n)).strip().split(' '))
mergeSort(arr)
print("Sorted array is ", *arr)

```

### Way-2 : Without using slicing

```python
# Python program for implementation of MergeSort
 
# Merges two subarrays of arr[].
# First subarray is arr[l..m]
# Second subarray is arr[m+1..r]

# Note- You need all the 3 numbers (first, mid and last)  to merge the 2 arrays : l =left , m =mid, r =right
def merge(arr, l, m, r):
    n1 = m - l + 1
    n2 = r - m
 
    # create temp arrays
    L = [0] * (n1)
    R = [0] * (n2)
 
    # Copy data to temp arrays L[] and R[]
    for i in range(0, n1):
        L[i] = arr[l + i]
 
    for j in range(0, n2):
        R[j] = arr[m + 1 + j]
 
    # Merge the temp arrays back into arr[l..r]
    i = 0     # Initial index of first subarray
    j = 0     # Initial index of second subarray
    k = l     # Initial index of merged subarray
 
    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1
 
    # Copy the remaining elements of L[], if there
    # are any
    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1
 
    # Copy the remaining elements of R[], if there
    # are any
    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1

        
# l is for left index and r is right index of the
# sub-array of arr to be sorted
def mergeSort(arr, l, r):
    if l < r:
 
        # Same as (l+r)//2, but avoids overflow for
        # large l and h
        m = l+(r-l)//2  # m is mid
 
        # Sort first and second halves
        mergeSort(arr, l, m)
        mergeSort(arr, m+1, r)
        merge(arr, l, m, r)
 
 
# Driver code to test above
arr = [12, 11, 13, 5, 6, 7]
n = len(arr)
print("Given array is")
for i in range(n):
    print("%d" % arr[i],end=" ")
    
mergeSort(arr, 0, n-1)
print("\n\nSorted array is")
for i in range(n):
    print("%d" % arr[i],end=" ")
```
