# 0/1 Knapsack

A knapsack is basically a bag. A bag has a maximum weight capacity. You are given few items with their weights and how valuable they are, you need to get the maximum value you can carry in that bag within the given weight constraint. ( Imagine going on a flight with luggage restriction of 15kg or going to trek with weight limits).<br>

You are given a weight array, a values array and the maximum weight that the knapsack can carry. You need to find the maximum value you can put in knapsack within the maximum weight.

```python
def knapsack(weights, values, n, maxWeight, i=0) :
    
    if i == n:
        return 0
    
    
    if weights[i] > maxWeight:
        
        ans = knapsack(weights, values, n, maxWeight, i+1)
        
    else:
        
        ans1 = values[i] + knapsack(weights, values, n, maxWeight - weights[i], i+1)
        
        ans2 = knapsack(weights, values, n, maxWeight, i+1)
        
        ans = max(ans1, ans2)
        
        
    return ans
```

```python
n = 4
weights = [1, 2, 4, 5]
values = [5, 4, 8, 6]
maxWeight = 5

knapsack(weights, values, n, maxWeight)
```
