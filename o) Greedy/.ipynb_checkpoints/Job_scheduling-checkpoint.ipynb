```python
def find_max_profit_from_jobs_2(jobs):

    jobs = sorted(jobs, key=cmp_to_key(my_comparator), reverse=True)
    
    n = len(jobs) # number of jobs
    
    # create a dp array to store intermediate results
    dp = [-1 for i in range(len(jobs)+1)]

    # if there are no jobs, then max-profit is 0
    # this acts as the base case for our bottom-up 1-D dp
    dp[0] = 0

    for i in range(1, n+1):

        max_profit_excluding_current_job = dp[i-1] # avoiding recursive call using dp array

        max_profit_including_current_job = jobs[i-1].profit

        non_overlapping_job_index = -1

        for j in range(i-1, 0, -1):
            # if the start time of current job (i) is greater than a previous job (j) then we can
            # include the max-profit of that previous job with the current job
            if jobs[j].end <= jobs[i-1].start:
                non_overlapping_job_index = j
                break

        # if we found a non overlapping job for this index then we can add its profit to current job profit
        if non_overlapping_job_index != -1:
            max_profit_including_current_job += jobs[j].profit

        dp[i] = max(max_profit_excluding_current_job, max_profit_including_current_job)

    return dp[n]

from functools import cmp_to_key

def my_comparator(a, b):
    # "b" points to first element of the list and "a" points to second element of the list
    # a = ['Cathy',22] and b = ['Bella',29]
    # when this function will be called for 1st time, a = ['Cathy',22] and b = ['Bella',29]
    # in comparator for ascending sort return -1 if a<b, 0 if a=b and 1 if a > b
    # b = first element
    # a = second element
    # return 1 means the elements are placed in CORRECT ASCENDING ORDER already i.e. first element should be first in sorted array only
    # similarly, -1 will indicate that the order of a and b must be reversed
    # IF YOU WANT DECENDING ORDER THEN RETURN THE OPPOSITE VALUES, i.e. -1 for 1 and 1 for -1
    # 0r Sort in ascending order only and then pass reverse=True
    
    if b.end < a.end:
        return 1
    elif a.end > b.end:
        return -1
    else:
        return 0

class Job_properties:

    def __init__(self, start, end, profit):

        self.start = start
        self.end = end
        self.profit = profit

def jobs_scheduling():
    jobs = [Job_properties(2,5,50), Job_properties(1,2,20), Job_properties(6,9,70), Job_properties(2,50,400)]

    max_profit = find_max_profit_from_jobs_2(jobs)

    print(max_profit)
```
