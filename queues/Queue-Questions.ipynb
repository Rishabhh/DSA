### Reverse a queue

You have been given a queue that can store integers as the data. You are required to write a function that reverses the populated queue itself without using any other data structures.

```python
import queue
from sys import setrecursionlimit

setrecursionlimit(10 ** 6)
def reverseQueue(inputQueue) :
    if inputQueue.qsize() <= 1 :
        return
    front = inputQueue.get()
    reverseQueue(inputQueue);
    inputQueue.put(front)
```

### Reverse the First K Elements in the Queue

**For a given queue containing all integer data, reverse the first K elements.**

NOTE - You have been required to make the desired change in the input queue itself. <br>





### Way-1 : 

Use recursion stack

```python
import queue

def reverseKElementsHelper(inputQueue, k):
    
    if k ==0 :
        return inputQueue
    
    element = inputQueue.get()
    inputQueue = reverseKElementsHelper(inputQueue, k-1)
    
    inputQueue.put(element)
    
    return inputQueue

def reverseKElements(inputQueue, k) :
    #Your code goes here
    
    inputQueue = reverseKElementsHelper(inputQueue, k)
    
    for i in range(inputQueue.qsize()-k):
        
        element = inputQueue.get()
        inputQueue.put(element)
      
    return inputQueue
```

```python
import queue

def reverseKElements(inputQueue, k) :
    if (inputQueue.empty()) or (k > inputQueue.qsize()) :
        return inputQueue
    
    if k <= 0 :
        return inputQueue
    
    stack = list()
    
    for i in range(k) :
        stack.append(inputQueue.get())
    
    while not isEmpty(stack) :
        inputQueue.put((stack.pop()))
    
    for i in range(inputQueue.qsize() - k) :
        inputQueue.put(inputQueue.get())
    
    return inputQueue
```
