## Queue using 2 stacks

Here one of the operations either Enqueue or Dequeue can be O(n) operations but not both


```python
class QueueUsingStacks:
    
    def __init__(self):
        
        self.__s1 = []
        self.__s2 = []
    
    # O(1)
    def enqueue(self, data):
        
        self.__s1.append(data)
    
    # O(n)
    def dequeue(self):
        
        if len(self.__s1) == 0:
            print("queue is empty !")
            return -1
        
        while len(self.__s1) != 0:
            
            element = self.__s1.pop()
            self.__s2.append(element)
            
        element = self.__s2.pop()
        
        while len(self.__s2) != 0:
            
            element = self.__s2.pop()
            self.__s1.append(element)
            
        return element
    
    def isEmpty(self):
        
        return self.size() == 0
    
    def size(self):
        
        return len(self.__s1)
    
    def front(self):
        
        if self.isEmpty():
            print("queue is empty !")
            return -1
        
        return self.__s1[0]
```

```python
q = QueueUsingStacks()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)

while q.isEmpty() is False:
    print(q.front())
    q.dequeue()
    
print(q.dequeue())
```

### In below implementation - Enqueue is O(n) and Dequeue is O(1)

Basically whenever enqueue is called -  we copy all the elements in s1 to s2 by pop-s1 and append-s2 and then we perform enqueue by inserting the new element in s1 and then copy the elements back from s2 to s1 by doing pop-s2 and append-s1.
