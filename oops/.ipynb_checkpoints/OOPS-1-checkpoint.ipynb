```python
class Solution:
    def subArraySum(self,arr, n, s):
        current_sum = 0
        start = 0
        end = 0

        while start<=end:
            
            print("X-CURRENT_SUM IS", current_sum)
            print("X-START IS", start)
            print("X-END IS", end)

            if current_sum == s:
                return [start+1, end+1]

            elif current_sum > s:
                current_sum = current_sum - arr[start]
                start = start + 1

            else:
                if end == n:
                    break
                current_sum = current_sum + arr[end]
                end = end+1
                
            print(current_sum)
            print("start is", start)
            print("end is", end)

        return -1
    
    
        
        
        
```
