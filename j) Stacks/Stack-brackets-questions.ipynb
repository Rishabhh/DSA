### Balanced Paranthesis

**For a given a string expression containing only round brackets or parentheses, check if they are balanced or not. Brackets are said to be balanced if the bracket which opens last, closes first.**

Expression: (()())
Since all the opening brackets have their corresponding closing brackets, we say it is balanced and hence the output will be, 'true'.





```python
def isBalanced(expression) :
	#Your code goes here
    
    s = []
    
    for c in expression:
        
        if c in "{[(":
            s.append(c)
        elif c in "}":
            if len(s)==0 or s[-1] != '{':
                return False
            s.pop()
        elif c in ")":
            if len(s)==0 or s[-1] != '(':
                return False
            s.pop()
        elif c in "]":
            if len(s)==0 or s[-1] != '[':
                return False
            s.pop()
        else:
            continue
    
    if len(s) == 0:
        return True
    else:
        return False
```

### Check redundant brackets

**For a given expression in the form of a string, find if there exist any redundant brackets or not. It is given that the expression contains only rounded brackets or parenthesis and the input expression will always be balanced.**

**A pair of the bracket is said to be redundant when a sub-expression is surrounded by unnecessary or needless brackets.**

Example: <br>
Expression: (a+b)+c <br>
Since there are no needless brackets, hence, the output must be 'false'.

Expression: ((a+b)) <br>
The expression can be reduced to (a+b). Hence the expression has redundant brackets and the output will be 'true'.

```python

def checkRedundantBrackets(expression) :
	# Your code goes here
    
    s = []
    for c in expression:
        
        if c == ')':
            
            count = 0
            
            while s.pop() != '(':
                count = count + 1
            
            # this basically means there is no expression 
            # like a+ something that this bracket pair contrains
            # compare ((a+b)) to (a+(b+c)) - 1st example has redundant bracket
            if count == 0 or count == 1: 
                
                return True
        
        else:
            s.append(c)
            
            
    return False
```

### Minimum bracket Reversal Count

**For a given expression in the form of a string, find the minimum number of brackets that can be reversed in order to make the expression balanced. The expression will only contain curly brackets.
If the expression can't be balanced, return -1.**

Example: <br>
Expression: {{{{ <br>
If we reverse the second and the fourth opening brackets, the whole expression will get balanced. Since we have to reverse two brackets to make the expression balanced, the expected output will be 2.<br>

Expression: {{{ <br>
In this example, even if we reverse the last opening bracket, we would be left with the first opening bracket and hence will not be able to make the expression balanced and the output will be -1.

```python
# Algorithm :

# part-1 :
# remove all the balanced part of the string
# -- add to stack if its opening bracket
# -- if it is a closing bracket then you need to think
# ---- can there be a pair {} made immediately so that that balanced bracket is removed
# ---- therefore check at the top of the stack if it is opening bracket : if yes just pop
# ---- if not then add } to the stack 

# part-2 :
# Now just unbalanced part will be left
# -- pop elements 2 at a time to check how many to reverse for balancing them
# -- if the 2 elements are c1 and c2 , then there can be 2 cases for c1 and c2
# ---- if c1 = c2  ex : {{ or }} -> in both cases you need to just reverse 1 bracket
# ---- if c1 != c2 : 
# ------- then there are 2 cases : 
# --------- { } -> this case would be already removed in part-1 in removing balanced part as it is balanced
# --------- } {  -> here we need to reverse both bracket to make it a balanced string, therefore 2 reversal

def countBracketReversals(inputString) :
    # Your code goes here
    
    stack = list()
    
    if len(inputString) %2 != 0:
        return -1
    
    for c in inputString:
        
        if c == '{':
            
            stack.append(c)
            
        else:
            
            if len(stack) !=0 and stack[-1] == '{':
                
                stack.pop()
            
            else:
                stack.append(c)
                
    
    
    count = 0
    
    while len(stack) != 0:
        
        c1 = stack.pop()
        c2 = stack.pop()
        
        if c1 == c2:
            
            count = count + 1
            
        elif c1 != c2:
            
            count = count + 2
            
    return count
```
